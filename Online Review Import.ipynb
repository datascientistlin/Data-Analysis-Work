```python
import json
import requests
import pandas as pd
from pandas.io.json import json_normalize
```

```python
#API key is removed in this file
headers = {'x-api-key': ''}
```

```python
review_sites = ['BestBuy/en-US', 'Amazon/en-US', 'Amazon/de-DE', 'Amazon/en-GB', 'Amazon/jp-JP']
```

```python
df = pd.DataFrame()
```

```python
for review_site in review_sites:
    url = 'https://customerreviewtracking-prod-we.azurewebsites.net/api/reviews/jabra/'+review_site
    print (url)
    response = requests.get(url, headers=headers)
    print (response.status_code)
    review_data = json_normalize(response.json(), 'reviews', ['productName'])
    review_data['review_site'] = review_site
    df = df.append(review_data, ignore_index=True)
    print(df.shape)
```

```python
df.head()
```

```python
df_sort = df.rename(columns={'body':'review_text', 
                             'isVerifiedPurchase':'verified_purchase', 
                             'rating':'review_rating',
                             'reviewId':'review_id', 
                             'reviewUrl':'review_url', 
                             'time':'review_date',
                             'title':'review_title',
                             'productName':'product_name'})
```

```python
df_sort.head()
```

```python
df_sort['review_date'] = df_sort['review_date'].astype('datetime64[ns]')
```

```python
df_sort.head()
```

```python
df_sort = df_sort[['review_site','product_name','review_id','review_date','review_rating','review_title','review_text','verified_purchase','review_url']]
```

```python
df_sort.head()
```

```python
df_sort.sort_values(by=['review_site','product_name','review_date'], inplace=True)
```
